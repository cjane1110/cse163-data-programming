## Task 1 format_event_dates

This task converts a list of date strings from a long format (e.g., "January 1st, 2000") into a numeric format (e.g., "1/1/2000").
If the input list is empty, it returns an empty list.

```python
def format_event_dates(date_list):
    """
    Convert dates like 'January 1st, 2000' into '1/1/2000'.

    Parameters:
        date_list (list): list of date strings

    Returns:
        list: reformatted date strings
    """
    month_to_number = {
        "January": 1,
        "February": 2,
        "March": 3,
        "April": 4,
        "May": 5,
        "June": 6,
        "July": 7,
        "August": 8,
        "September": 9,
        "October": 10,
        "November": 11,
        "December": 12
    }

    formatted_dates = []

    for date in date_list:
        parts = date.split()
        month = parts[0]
        day = parts[1].replace(",", "")
        year = parts[2]

        if day.endswith("st") or day.endswith("nd") or day.endswith("rd") or day.endswith("th"):
            day = day[:-2]

        month_number = month_to_number[month]
        formatted_dates.append(f"{month_number}/{int(day)}/{year}")

    return formatted_dates
```

```python
sample_dates = [
    "January 1st, 2000",
    "February 22nd, 1999",
    "March 3rd, 2024",
    "April 14th, 2026"
]

print(format_event_dates(sample_dates))
```

```python
assert format_event_dates(["January 1st, 2000"]) == ["1/1/2000"]
assert format_event_dates(["February 22nd, 1999"]) == ["2/22/1999"]
assert format_event_dates(["March 3rd, 2024", "April 14th, 2026"]) == ["3/3/2024", "4/14/2026"]
assert format_event_dates([]) == []
```

## Task 2 make_coordinate_pairs

Given two lists of x-values and y-values, return a list of coordinate
pairs as tuples. Each pair combines values at the same index.

If the lists are different lengths, stop at the shorter list.

```python
def make_coordinate_pairs(x_list, y_list):
    """
    Combine two lists into a list of (x, y) tuples.

    Parameters:
        x_list (list): list of x values
        y_list (list): list of y values

    Returns:
        list: list of (x, y) tuples
    """
    result = []

    # use the shorter length to avoid errors
    length = min(len(x_list), len(y_list))

    for i in range(length):
        pair = (x_list[i], y_list[i])
        result.append(pair)

    return result
```

```python
x = [0, 1, 2]
y = [3, 4, 5]

print(make_coordinate_pairs(x, y))
```

```python
assert make_coordinate_pairs([1, 2, 3], [4, 5, 6]) == [(1, 4), (2, 5), (3, 6)]
assert make_coordinate_pairs([1], [9]) == [(1, 9)]
assert make_coordinate_pairs([], []) == []
assert make_coordinate_pairs([1, 2], [3]) == [(1, 3)]
```
