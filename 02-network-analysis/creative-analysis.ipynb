This notebook models illustrative courses and prerequisite relationships with a directed graph. The example was manually authored for this portfolio; it is not official curriculum data and does not use external social-network or user data.

```python
import networkx as nx


class CourseNetwork(nx.DiGraph):
    """A directed graph representing course prerequisites."""

    def add_course(
        self,
        course_id: str,
        title: str,
        department: str,
        credits: int,
        difficulty: str,
    ) -> None:
        """Add a course node and its descriptive attributes."""
        self.add_node(
            course_id,
            title=title,
            department=department,
            credits=credits,
            difficulty=difficulty,
        )

    def add_prerequisite(self, prereq: str, course: str) -> None:
        """Add a directed edge from a prerequisite to a course."""
        self.add_edge(prereq, course)

    def get_prerequisites(self, course: str) -> list[str]:
        """Return the direct prerequisites for a course."""
        return list(self.predecessors(course))

    def get_next_courses(self, course: str) -> list[str]:
        """Return courses that directly depend on a course."""
        return list(self.successors(course))

    def get_course_info(self, course: str) -> dict:
        """Return the stored attributes for a course."""
        return dict(self.nodes[course])
```

```python
courses = CourseNetwork()

courses.add_course("CSE121", "Programming I", "CSE", 4, "beginner")
courses.add_course("CSE122", "Programming II", "CSE", 4, "beginner")
courses.add_course(
    "CSE163",
    "Intermediate Data Programming",
    "CSE",
    4,
    "intermediate",
)

courses.add_prerequisite("CSE121", "CSE122")
courses.add_prerequisite("CSE122", "CSE163")

print(courses.get_prerequisites("CSE163"))
print(courses.get_next_courses("CSE121"))
print(courses.get_course_info("CSE163"))
```

## Approach

A directed graph fits prerequisite data because each relationship has a clear direction: a prerequisite leads to a later course. Course nodes store title, department, credit, and difficulty information. The helper methods make it easy to add courses, connect prerequisites, inspect requirements, find follow-on courses, and retrieve course details. This structure could support degree planning or prerequisite validation.
