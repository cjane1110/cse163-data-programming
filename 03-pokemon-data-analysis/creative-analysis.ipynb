## Filling a missing attack value

I am filling each missing attack (`atk`) value with the average attack for that row's own Pokémon type. I use pandas `groupby()` and `transform("mean")` so every row is aligned with its type mean before `fillna()` replaces missing values. If a type has no valid attack values, as with flying Pokémon in this dataset, its mean is unavailable and the attack remains missing rather than borrowing a value from another type.

```python
import pandas as pd


def fill_missing_attack(pokemon: pd.DataFrame) -> pd.DataFrame:
    """
    Fill missing attacks using the mean attack for each row's type.
    """
    type_means = pokemon.groupby("type")["atk"].transform("mean")
    pokemon["atk"] = pokemon["atk"].fillna(type_means)

    return pokemon
```

```python
original = pd.read_csv("pokemon_missing.csv")
attack_pokemon = fill_missing_attack(original.copy())

dragon_mean = original.loc[original["type"] == "dragon", "atk"].mean()
dragonair_attack = attack_pokemon.loc[
    (attack_pokemon["name"] == "Dragonair")
    & (attack_pokemon["level"] == 13),
    "atk",
].iloc[0]
flying_mean = original.loc[original["type"] == "flying", "atk"].mean()
staravia_attack = attack_pokemon.loc[
    attack_pokemon["name"] == "Staravia", "atk"
].iloc[0]

assert dragonair_attack == dragon_mean == 54.2
assert pd.isna(staravia_attack) and pd.isna(flying_mean)
attack_pokemon[attack_pokemon["atk"].isna()]
```
